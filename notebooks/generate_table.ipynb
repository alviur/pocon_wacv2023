```python
dfs = []

for filename in files:
    df = pd.read_csv(filename, sep='\t', names=[f'Task {i+1:02d}' for i in range(10)])
    d = filename.split('/')[-3]    
    s = d.split('_')
    ordering_name = '_'.join(s[:-4])
    df['method'] = methods[s[-4]]['name']
    df['ordering'] = splits[ordering_name]['name']
    dfs.append(df)

df = pd.concat(dfs, axis=0, ignore_index=True)
```

#### only task2, task5, task10

```python
TASKS_NAMES = ['Task 02', 'Task 05', 'Task 10']
DIMS = ['method', 'ordering']
df1 = df[DIMS + TASKS_NAMES]
```

```python
df2 = df1.melt(DIMS, TASKS_NAMES, var_name='task')
```

```python
df2.groupby('method').size()
```

```python
df2.groupby(['method', 'task']).size()
```

```python
tab = df2.pivot_table(index=['method', 'task'], columns=['ordering'], values=['value'])
tab2 = tab * 100
```

```python
tab2.sort_index(level=[0,1], ascending=[True, True], inplace=True)
tab2
```

```python
tab2 = tab2[[
    ('value',     'Random'),
    ('value', 'iCaRL seed'),   
    ('value',  'Max Conf.'),
    ('value',  'Min Conf.'),
    ('value',    'CM Dec.'),
    ('value',     'CM Eq.'),
    ('value',    'CM Inc.'),
     ]]
```

```python
max_values = tab2.max(axis=1)
min_values = tab2.min(axis=1)
```

```python
latex = tab2.to_latex(float_format='%.1f')

lines = []
i = 0
for l in latex.split('\n'):
    if 'Task' in l:
        # result line
        v = f'{max_values.values[i]:0.1f}'
        l = l.replace(v, "\\textbf{" + v + "}")
        v = f'{min_values.values[i]:0.1f}'
        l = l.replace(v, "\\underline{" + v + "}")
        i += 1
    else:
        l = l.replace('value', '')
        l = l.replace('ordering', '')
    lines.append(l)
    
latex = '\n'.join(lines)
print(latex)
```
